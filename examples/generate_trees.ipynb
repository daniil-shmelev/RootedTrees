```python
import rootedtrees as rt
```

## Next tree with respect to lexicographic order

```python
t = rt.Tree([[[[]],[]],[[]]])
rt.display(t)
print("Next tree:")
rt.display(rt.next_tree(t))
```

## All trees of a given order

```python
for n in range(6):
    print("Trees of order " + str(n) + ":")
    for t in rt.trees_of_order(n):
        rt.display(t)
```

## All trees up to a given order

```python
print("All trees up to order 4:")
for t in rt.trees_up_to_order(4):
    rt.display(t)
```
